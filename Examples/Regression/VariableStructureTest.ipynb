# Importing Libraries and Writing Test Functions

```python
from learntools import Network, Learning, VariableStructure
import matplotlib.pyplot as plt
import numpy as np
```

```python
def linearTestData(n):
    x = np.arange(0,n)/n
    y = np.arange(0,n)/n
    x = [[i] for i in x]
    y = [[i] for i in y]
    return x,y

def linearNormalTestData(n):
    x = np.arange(0,n)/n
    y = np.arange(0,n)/n + np.random.normal(size=n)/n
    x = [[i] for i in x]
    y = [[i] for i in y]
    return x,y

def quadraticTestData(n):
    x = np.arange(0,n) /n
    y = np.arange(0,n)**2/n**2
    x = [[i] for i in x]
    y = [[i] for i in y]
    return x,y

def sinTestData(n):
    x = np.arange(0,n)
    y = (np.sin(2*np.pi*x/(n))+1)/2
    x = [[i] for i in x/n]
    y = [[i] for i in y]
    return x,y

def MSELoss(y,ypred):
    return np.sum((y-ypred)**2)

LossFunction = lambda net: MSELoss(y,net.forward(x))
```

# Quick tests on a few functions

```python
x,y = linearNormalTestData(10)
plt.plot(x,y,'o')
```

```python
x,y = linearTestData(100)
netLinear, losses = VariableStructure.VariableStructure(1,1,LossFunction,Network.relu,info = True)

print('Final Loss',losses[-1])
```

```python
#plots
plt.subplot(1, 2,1)
plt.plot(losses)
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.yscale('log')

# Plot 2
plt.subplot(1, 2, 2)
plt.plot(x, y, 'o', label='Data')
plt.plot(x, netLinear.forward(x), label='Network Prediction')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data and Network Prediction')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
x,y = quadraticTestData(100)
netQuad, losses = VariableStructure.VariableStructure(1,1,LossFunction,Network.relu,info = True)

print('Final Loss',losses[-1])
```

```python
#plots
plt.subplot(1, 2,1)
plt.plot(losses)
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.yscale('log')

# Plot 2
plt.subplot(1, 2, 2)
plt.plot(x, y, 'o', label='Data')
plt.plot(x, netQuad.forward(x), label='Network Prediction')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data and Network Prediction')
plt.legend()

plt.tight_layout()
plt.show()
```
