```python
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
from backtesting.test import SMA, GOOG
from learntools import Learning, Network
#from backtesting import set_bokeh_output
#set_bokeh_output(notebook=False)

class SmaCross(Strategy):
    n1 = 10
    n2 = 20
    def init(self):
        close = self.data.Close
        self.sma1 = self.I(SMA, close, self.n1)
        self.sma2 = self.I(SMA, close, self.n2)

    def next(self):
        if crossover(self.sma1, self.sma2):
            self.buy()
        elif crossover(self.sma2, self.sma1):
            self.sell()


bt = Backtest(GOOG, SmaCross,
              cash=10000, commission=.002,
              exclusive_orders=True)

output = bt.run()
print(output)
bt.plot()
```

```python
import numpy as np
from learntools import Network, Learning
class LinearStrat(Strategy):


    def init(self):
        close = self.data.Close
        n = len(close)
        self.MA10 = np.mean(close[-min(100,n):])
        self.MA20 = np.mean(close[-min(200,n):])
        

    def next(self):
        data = [self.MA10,self.MA20]
        #print(data)
        
        #print(len(input))
        out = self.net.forward([data])

        if out[0][0] > 0:
            self.buy()
        else:
            self.sell()

net = Network.network(2,1)
net.add_layer(Network.layer_dense(2,2))
net.add_layer(Network.relu())
net.add_layer(Network.layer_dense(2,1))
net.add_layer(Network.sigmoid())
#net.random_initilisation()

def run(strat, net):
    strat.net = net
    bt = Backtest(GOOG, strat,
                  cash=10000, commission=.002,
                  exclusive_orders=True)
    
    output = bt.run()
    return output, bt

run(LinearStrat,net)
```

```python
loss = lambda net: np.exp(-run(LinearStrat,net)[0]['Return [%]'])
```
